```python
import pandas as pd
import mysql.connector
```

```python
 connection = mysql.connector.connect(host='localhost',
                                         database='nrif_analysis',
                                         user='<mysqlserver_username>',
                                         password='<YourPassword>')
```

```python
pd_institutes_data = pd.read_csv("NRIF_Analysis.csv")
institutes_df = pd.DataFrame(pd_institutes_data)
institutes_df
```

```python
mySql_Create_Table_Query = """CREATE TABLE institutes (Institute_Id varchar(11),Institute_Name varchar(250),nrif_rank INT) """
cursor = connection.cursor()
result = cursor.execute(mySql_Create_Table_Query)
```

```python
for row in institutes_df.itertuples():
    mySql_insert_Query = "INSERT INTO institutes (Institute_Id,Institute_Name,nrif_rank) VALUES(%s, %s,%s)"
    val=(row[1],row[2],row[3])
    cursor = connection.cursor()
    result = cursor.execute(mySql_insert_Query,val)
connection.commit()
```

```python
pd_intake_data = pd.read_csv("Intake_Placements.csv")
intake_df = pd.DataFrame(pd_intake_data)
intake_df
```
